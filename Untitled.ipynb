```python
import pandas as pd
import numpy as np
import io
import os
```

```python
df=pd.read_csv(r"C:\Users\swasa\Desktop\adult.csv") #downloded the dataset from given url and loaded in Jupyternotebook
```

```python
df.head(20) # will display a summary statistics of all observed features and labels
```

```python
#finding missing data

#replace all the "?" to NA
df = df.replace("?",np.NaN)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#cols_mode = ['workclass', 'occupation', 'native.country']
for column in ['workclass', 'occupation', 'native.country']:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
# label encoding the data 
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
  
df['workclass']= le.fit_transform(df['workclass']) 
df['education']= le.fit_transform(df['education']) 
df['marital.status']= le.fit_transform(df['marital.status']) 
df['occupation']= le.fit_transform(df['occupation']) 
df['relationship']= le.fit_transform(df['relationship']) 
df['race']= le.fit_transform(df['race'])
df['sex']= le.fit_transform(df['sex'])
df['native.country']= le.fit_transform(df['native.country'])
```

```python
y=df.income
X=df.drop('income',axis=1) # Extracting X as all columns except the Income column and Y as Income column. 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train.head()
```

```python
X_train.shape      # to conform the no. of colums in the train_data set
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
#Create a Gaussian Classifier
gnb = GaussianNB()
# import Multinomial Naive Bayes model
from sklearn.naive_bayes import MultinomialNB
#instantiate a Multinomial Naive Bayes model
nb = MultinomialNB()
#Train the model using the training sets
gnb.fit(X_train, y_train)
nb.fit(X_train, y_train)
```

```python
%time nb.fit(X_train, y_train)
#Predict the response for test dataset
y_pred = gnb.predict(X_test)
# make class predictions for test dataset
y_pred_class = nb.predict(X_test)

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy of Gaussian Classifier",metrics.accuracy_score(y_test, y_pred))
# Model Accuracy of Multinomial Naive Bayes 
print("Accuracy of Multinomial Naive Bayes",metrics.accuracy_score(y_test, y_pred_class))
# print the confusion matrix of Gaussian Naive Bayes 
print("Accuracy of Gaussian Naive Bayes\n",metrics.confusion_matrix(y_test, y_pred))
# print the confusion matrix
print("Accuracy of Multinomial Naive Bayes\n",metrics.confusion_matrix(y_test, y_pred_class))
```
